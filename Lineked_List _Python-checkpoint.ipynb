```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
def emptyList():
    raise Exception("Linked List is empty")
    
class LinkedList:
    def __init__(self):
        self.head=None
        
    def insert_at_beginning(self,data):
        for li in data:
            node=Node(li,self.head)
            self.head=node
            
    def insert_at_end(self,data):
        temp=Node(data)
        if self.head==None:
            self.head=temp
        else:
            curr=self.head
            while curr.next!=None:
                curr=curr.next
            curr.next=temp
    
    def list_length(self):
        count=0
        itr=self.head
        while itr:
            count+=1
            itr=itr.next
        print("Length of List is:",count)
    
    def delete_at(self,index):
        if index<0:
            raise Exception("Invalid Index")
        if index==0:
            self.head=self.head.next
            return
        count=0
        itr=self.head
        while itr:
            if count== index-1:
                itr.next=itr.next.next
            itr=itr.next
            count+=1
    
    def delete_at_beginneing(self):
        if self.head is None:
            self.emptyList()
        self.head=self.head.next
        
    def delete_at_end(self):
        if self.head==None:
            self.emptyList()
        else:
            curr=self.head
            prev=None
            while curr.next!=None:
                prev=curr
                curr=curr.next
            prev.next=curr.next
            del curr
            
    def add_item_anywhere(self,index,data):
        if int(index)<0:
            raise Exception("Invalid Index")
        else:
            itr=self.head
            count=0
            while itr:
                if count == index-1:
                    node=Node(data,itr.next.next)
                    itr.next=node
                itr=itr.next
                count+=1
    def Print(self):
        if self.head is None:
            print("linked list is empty")
        itr=self.head
        llitems=""
        while itr:
            llitems+=str(itr.data)+"-->"
            itr=itr.next
        print(llitems)

li=LinkedList()
week=["Sunday","Monday","Tuesday","Wedensday","Thursday","Friday","Saturday"]
print("Insert items at Beginning")
li.insert_at_beginning(week)
li.insert_at_end("Hii")
li.Print()
li.list_length()
li.delete_at(6)
li.Print()
print("delete item at beginning")
li.delete_at_beginneing()
li.Print()
print("delete Item at end")
li.delete_at_end()
li.Print()
li.add_item_anywhere(4,"TT")
li.Print()
li.list_length()
```
